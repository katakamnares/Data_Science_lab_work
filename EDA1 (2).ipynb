# importing the all libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Cardiotocographic.csv")
df.head()
```

# Data Cleaning

```python
# checking the dtypes
df.dtypes
```

```python
# checking the info of the Dataset
df.info()
```

```python
# Checking the Missing Values
df.isnull().sum()
```

```python
# Handiling the missing Values
df["LB"]=df["LB"].fillna(df["LB"].median())
df["AC"]=df["AC"].fillna(df["AC"].median())
df["DS"]=df["DS"].fillna(df["DS"].median())
df["DP"]=df["DP"].fillna(df["DP"].median())
df["MLTV"]=df["MLTV"].fillna(df['MLTV'].median())
df["Width"]=df["Width"].fillna(df["Width"].median())
df["Tendency"]=df["Tendency"].fillna(df["Tendency"].median())
df["NSP"]=df["NSP"].fillna(df["NSP"].median())
```

# Satstical Summary

```python
# calculating a stastical Summary
df.describe()
```

```python
# calculating the median
df.median()
```

```python
# Calculating the IQR of all columns
from collections import defaultdict
Dic={}
col=df.columns[:]
for x in col:
  q1=df[x].quantile(0.25)
  q3=df[x].quantile(0.75)
  IQR=q3-q1
  Dic[x]=IQR
Dic
```

#### In these columns LB,AC,DS,DP,MLTV,Width,Tendency,NSP outliers are not present.
