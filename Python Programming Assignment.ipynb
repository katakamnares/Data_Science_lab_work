# Prime Numbers

```python
# Asking the user input number
num=int(input())
#finding the factors for given number
list=[i for i in range(2,num+1) if num % i==0]
# checking the whether given number is prime number or not
if len(list)==1:
  print(" Given Number is Prime Number")
else:
  print("Given Number is Not Prime Number")
```

# Product of Numbers

```python
#importing numpy for random numbers
import numpy as np
# Generating The random Numbers
num=np.random.randint(1,100,2)
#Asking the user for the answer
answer=input(f"Enter the Product of {num[0]} * {num[1]} is:")
result=num[0]*num[1]
#Checking the user answer is correct or not
if result==answer:
  print("Your Answer is Correct")
else:
  print("Your Answer is not Correct")
```

# Squares of Even/Odd Numbers

```python
# here i choose the even numbers
#range is given 100 to 200
for i in range(100,200):
  # checking the even or odd
  if i %2==0:
    print(i**2)
```

# Word Counter

```python
# asking the user input and converting to list of strings
text=input().split()
text2={}
# Adding the count elements
for i in text:
  if i not in text2:
    text2[i]=1
  else:
    text2[i]+=1
# printing the all elements with count
for key in text2:
  print(f'{key}:{text2[key]}')
```

# Check for Palindrome

```python
# defining a function for palindrome
def is_palindrome(x):
  y=x[::-1]
# checking with given value and reversed value
  if x==y:
    return True
  else:
    return False
x=input()
# calling the function
print(is_palindrome(x))
```
