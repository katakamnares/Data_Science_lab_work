# Importing the libraries

```python
# importing the all libraries
import pandas as pd
from scipy.stats import chi2_contingency,norm
```

```python
# Intially we are collecting the Data
Smart_Thermo=[50,80,60,30,20]
Smart_Light=[70,100,90,50,50]
Stastification_Level=["Very Satisfied","Satisfied","Neutral","Unsatisfied","Very Unsatisfied"]
Dic={"SmartThermostats":Smart_Thermo,
     "Smart Lights":Smart_Thermo,
     "Stastification":Stastification_Level}
```

```python
Data=pd.DataFrame(Dic).set_index("Stastification")
Data
```

# Calculating the Chi_Square Test

```python
# Null Hypothesis(Ho):Thers is no Significant association between Types of purchased Home Smart Devices and Customer Stastification Level
# Alternative Hypothesis(H1):Thers is Significant association between Types of purchased Home Smart Devices and Customer Stastification Level.
alpha=0.05
# Applying Chi_Square_Statistics
Value=chi2_contingency(Data)
Cal_value=Value[0]
P_Value=Value[1]
P_Value
Cal_value
```

```python
# Calculating Critical Value
critical_value=norm.ppf(1-(alpha/2))
```

```python
# Decisiding Hypothesis
if Cal_value < critical_value:
  print("Accept Ho,Thers is no Significant association between Types of purchased Home Smart Devices and Customer Stastification Level")
else:
  print("Reject Ho,Thers is Significant association between Types of purchased Home Smart Devices and Customer Stastification Level")
```
