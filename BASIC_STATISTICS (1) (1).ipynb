# Standardization

#### standardization is a technique used to convert the numeric columns with different ranges into one range by using the formula
Z=x-mu/sigma


```python
import numpy as np
from matplotlib import pyplot as plt
#"Volume","Avg Price","Total Sales Value","Discount Rate(%)","Discount Amount","Net Sales Value"
df1=df.select_dtypes(include=np.number)
df2=df1.apply(lambda X:(X-X.mean())/X.std())
df2.hist()
plt.show()
df.iloc[:,1:].hist()
plt.show()
```

#### The distribution of the each column is same before and after standardization

# Conversion of Categorical Data into Dummy Variables

####While Traning,machine Learning models can't understand the categorical data.so that categorical data is converted into Binary format(0's and 1's)

```python
from sklearn.preprocessing import OneHotEncoder
df["Date"]=pd.DatetimeIndex(df.Date)
df1=df.select_dtypes(include=object)
cat_cols=df1.columns
one_he=OneHotEncoder()
one=one_he.fit_transform(df.select_dtypes(include=object)).toarray().astype(int)
# print(one_he.get_feature_names_out())
df3=pd.DataFrame(one,columns=one_he.get_feature_names_out())
# df3=df.drop(["Day","Brand","BU","SKU","City","Model"],axis=1)
# df3.join(df4)
df3

```
