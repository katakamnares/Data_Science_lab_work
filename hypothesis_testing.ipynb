```python
import pandas as pd
import numpy as np
from scipy import stats # Uncomment and specify what to import from scipy.stats
```

```python
# H0:μ = (W = $1,000 + $5X for X = 600 units)
# H1:μ > (W = $1,000 + $5X for X = 600 units)
# sample size < 30 then i used t-test statistic
alpha=0.05
x_bar=3050
n=25
s=125
mu=4000

t_cal=(x_bar-mu)/(s/np.sqrt(n))
print(t_cal)
```

```python
# obtaining the critical Value from the stats.norm.ppf function
t_crit=stats.norm.ppf(1-(alpha/2))
print(t_crit)
```

```python
# Comparing the t_cal value with the t_critical value
if t_cal < t_crit:
  print("Accept the null hypothesis(H0) and action is not to be taken")
else:
  print("Accept the Alternative hypothesis(H1) and action is to be taken")
```

# Conclusion

### Based on the above comparsion, null hypothesis is accepted. so that we conclude that there is no strong evidence to support the restaurant owners' claim that the weekly operating costs are higher than the model suggests.
